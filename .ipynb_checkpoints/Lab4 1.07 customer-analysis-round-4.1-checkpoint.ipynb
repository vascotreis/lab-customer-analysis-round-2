```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('cleandata.csv') 
```

```python
data
```

```python
print("shape", data.shape)
```

```python
data.dtypes
```

```python
data.apply(lambda s: pd.to_numeric(s, errors='coerce').notnull().all()) # check wich ones are numerical
```

```python
data.select_dtypes(exclude=['object']).columns.tolist() # numerical data
```

```python
data.select_dtypes(include=['object']).columns.tolist() # categorical data
```

```python
numerical = pd.DataFrame(data.select_dtypes(np.number))
numerical.head(1)
```

```python
categorical = pd.DataFrame(data.select_dtypes(object))
categorical.head(1)
```

```python
for i in numerical.columns:
    sns.displot(x = i, data = numerical)
    plt.show()
```

```python
for i in numerical.columns:
    plt.hist(x = data[i], density = False)  # density=False would make counts
    plt.ylabel('Count')
    plt.xlabel(i)
    plt.show()
```
