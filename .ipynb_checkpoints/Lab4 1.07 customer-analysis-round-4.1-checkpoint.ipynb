# Round 6

```python
categorical= categorical.drop(columns=['customer'],axis=1)
```

```python
encoder = OneHotEncoder().fit(categorical)
print(encoder.categories_)
cat_encoded = encoder.transform(categorical).toarray()
print(cat_encoded)
cols = encoder.get_feature_names_out(input_features=categorical.columns)
onehot_encoded = pd.DataFrame(cat_encoded,columns=cols)
onehot_encoded.head(5)
```

```python
onehot_encoded.head(5)
```

```python
data_normalised = x_normalised_dataframe.corr()
```

```python
X = pd.concat([x_normalised_dataframe, onehot_encoded], axis=1)
```

```python
display(X,y)
```

# Train and Run Model

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test,y_train, y_test=tts(X, y, test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression as LinReg

linreg=LinReg()    # model
linreg.fit(X_train, y_train)   # model training
y_pred_linreg=linreg.predict(X_test)   # model prediction
```

# Validation R2 scores

```python
print ('train R2: {} -- test R2: {}'.format(linreg.score(X_train, y_train),
                                            linreg.score(X_test, y_test)))
```

```python
from sklearn.metrics import mean_squared_error as mse


train_mse=mse(linreg.predict(X_train), y_train)
test_mse=mse(linreg.predict(X_test), y_test)

print ('train MSE: {} -- test MSE: {}'.format(train_mse, test_mse))
```
